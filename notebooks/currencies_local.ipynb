## Currency Pipeline — Local Explorer

Reads Bronze and Silver Delta Lake tables from the local `data/` directory.
Requires the pipeline to have been run with `RUNNING_LOCAL=True` in `.env`.
All cells are safe to run even if the directories do not exist yet.

```python
import os

import pandas as pd


def read_local(path: str) -> pd.DataFrame | None:
    """Read a local Delta Lake table as a pandas DataFrame.
    Returns None with a message if the directory does not exist.
    """
    if not os.path.exists(path):
        print(f"No data at {path}")
        return None
    return pd.read_parquet(path)
```

### Bronze — currencies

```python
bronze_currencies = read_local("../data/bronze/currencies")
if bronze_currencies is not None:
    print(f"{len(bronze_currencies):,} rows")
    display(bronze_currencies)
```

### Bronze — rates

```python
bronze_rates = read_local("../data/bronze/rates")
if bronze_rates is not None:
    print(f"{len(bronze_rates):,} rows")
    display(bronze_rates.head(20))
```

### Silver — currencies

```python
silver_currencies = read_local("../data/silver/currencies")
if silver_currencies is not None:
    print(f"{len(silver_currencies):,} rows")
    display(silver_currencies)
```
